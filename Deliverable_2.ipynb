```python
#Question two:2.	Write a Python script to create a plot comparing the trend in the number of cases of malaria over the years 2000 to 2021 in 6-11 month old males in Zambia, Zimbabwe, Mozambique, and Malawi. Prepare the plot as if it was going to be included in a publication.
```

```python
%matplotlib inline
rcParams['figure.figsize']=(14,7)
sns.set_style('whitegrid')
```

```python
malaria_cases_southernafrica_plot = malaria_cases[
    (malaria_cases['sex'] == 'Male') &
    (malaria_cases['age_group_name'] == '6-11 months') &
    (malaria_cases['cause_name'] == 'Malaria') &
    (malaria_cases['measure'] == 'prevalence') &
    (malaria_cases['metric'] == 'Number') &
    (malaria_cases['year'].between(2000, 2021)) &
    (malaria_cases['location_name'].isin(['Zambia', 'Zimbabwe', 'Mozambique', 'Malawi']))
]
```

```python
malaria_cases_southernafrica_plot
```

```python
for location, group_data in malaria_cases_southernafrica_plot.groupby('location_name'):
    plt.plot(group_data['year'], group_data['value'], label=location, marker='*')
plt.title('Malaria cases in Zambia, Zimbabwe, Mozambique, and Malawi among 6-12 month old males betwen 2000 and 2021')
plt.xlabel('Year')
plt.ylabel('Malaria cases')
plt.legend(title='Country')
plt.tight_layout
plt.show()
```

```python
#Question 3:3.	Use the provided data set of "draws" (which are samples, or possible values of the statistic being estimated) to estimate the prevalence of malaria in Zambia, Zimbabwe, Mozambique, and Malawi combined for 6-11 month old males in 2021, including 95% uncertainty intervals.


```

```python
population=pd.read_csv('populations.csv')
population
```

```python
sample=pd.read_csv('draws.csv')
sample
```

```python
population.columns.unique()
```

```python
sample.columns.unique()
```

```python
common_columns = sample.columns.intersection(population.columns)
print(common_columns)
```

```python
merged= pd.merge(sample, population, on=['age_group_name', 'location_name', 'sex', 'age_group_id', 'location_id',
       'sex_id', 'year_id'], how='left')
merged
```

```python
draws_cols = [col for col in merged.columns if col.startswith('draw')]
draws_cols
```

```python
merged[draws_cols]
```

```python
merged[draws_cols]=merged[draws_cols].mul(merged.population,axis=0)
merged
```

```python
total_draws=merged[draws_cols].sum()
total_draws
```

```python
total_population = merged.population.sum()
total_population
```

```python
prevalence = total_draws/total_population
prevalence
```

```python
prevalence.describe()
```

```python
lower = prevalence.quantile(0.025)
upper = prevalence.quantile(0.975)
uncertainity_interval = (lower, upper)
uncertainity_interval
```

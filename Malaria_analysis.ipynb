Q1: 1.	Write a Python script that creates a data set containing (only) the number of cases of malaria in 6-11 month old males in Zambia, Zimbabwe, Mozambique, and Malawi in 2021.

```python
# Import libraries and load CSV file for test data
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

# load the CSV file and name it malaria_data
malaria_data = pd.read_csv('test-data.csv')
malaria_data.head()
```

```python
# Confirm column names
print(malaria_data.columns)
```

```python
#confirm age_group_name
malaria_data.age_group_name.unique()
```

```python
malaria_data.metric.unique()
```

```python
#filter for 6-11 month old males by year(2021) and the 4 target countries
filtered_malaria_data = malaria_data[(malaria_data['age_group_name']=='6-11 months') &
                                   (malaria_data['sex'] =='Male') &
                                   (malaria_data['year'] == 2021) &
                                   (malaria_data['cause_name'] == 'Malaria') &
                                   (malaria_data['metric'] == 'Number') &
                                   (malaria_data['location_name'].isin(['Zambia', 'Zimbabwe', 'Mozambique','Malawi']))]
filtered_malaria_data.head()
```

The output containing (only) the number of cases of malaria in 6-11 month old males in Zambia, Zimbabwe, Mozambique, and Malawi in 2021

Question 2
Write a Python script to create a plot comparing the trend in the number of cases of malaria over the years 2000 to 2021 in 6-11 month old males in Zambia, Zimbabwe, Mozambique, and Malawi. Prepare the plot as if it was going to be included in a publication.


```python
# Check for the years required for the period of analysis
malaria_data.year.unique()
```

```python
filtered_malaria_data_new = malaria_data[(malaria_data['age_group_name']=='6-11 months') &
                                   (malaria_data['sex'] =='Male') &
                                   (malaria_data['year'] >= 2000) &
                                   (malaria_data['year'] <= 2021) &
                                   (malaria_data['cause_name'] == 'Malaria') &
                                   (malaria_data['metric'] == 'Number') &
                                   (malaria_data['location_name'].isin(['Zambia', 'Zimbabwe', 'Mozambique','Malawi']))]
filtered_malaria_data_new
```

```python
from matplotlib import pyplot as plt
#set default plot size and resolution (height, weight and DPI at 12,6 and 100 respectively
plt.rcParams['figure.figsize']= (12,6)
#use 100 DPI- high quality for publication
plt.rcParams['figure.dpi'] = 100

for location, group_data in filtered_malaria_data_new.groupby('location_name'):
    plt.plot(group_data['year'], group_data['value'], label=location)

plt.xlabel('Year')
plt.ylabel('Malaria Cases')
plt.title('Malaria Cases Over Time by country')
plt.legend()
plt.show()

```

```python
#plot using seaborn
import seaborn as sns
```

```python
# Set the size and resolution
plt.figure(figsize=(12, 6), dpi=100)

sns.lineplot(
    data=filtered_malaria_data_new,
    x='year',
    y='value',
    hue='location_name')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Malaria Cases')
plt.title('Malaria Cases among Males Over Time by Country ( 2000-2021)')
plt.tight_layout()
plt.show()
```

Q3. Use the provided data set of "draws" (which are samples, or possible values of the statistic being estimated) to estimate the prevalence of malaria in Zambia, Zimbabwe, Mozambique, and Malawi combined for 6-11 month old males in 2021, including 95% uncertainty intervals.


```python
#load both populations and draws data files
populations=pd.read_csv('populations_data.csv')
populations.head()
```

```python
draws=pd.read_csv('draws_data.csv')
draws.head()
```

```python
#location filters
draws.location_name.unique() 
```

```python
draws.columns.unique()
```

```python
populations.columns.unique()
```

```python
#combine both datasets
merged_data = pd.merge(draws, populations, 
                  on=['location_name', 'age_group_name'], how='left')
merged_data.head()
```

```python
# identify draws columns
draws_columns = [col for col in merged_data.columns if col.startswith('draw')]
draws_columns
```

```python
merged_data[draws_columns]
```

```python
merged_data[draws_columns]=merged_data[draws_columns].mul(merged_data.population,axis=0)
merged_data
```

```python
#total draws cases
summed_draws=merged_data[draws_columns].sum()
summed_draws
```

```python
#total population across countries
total_population = merged_data.population.sum()
total_population
```

```python
Overal_4_country_prevalence = summed_draws/total_population
Overal_4_country_prevalence
```

```python

Overal_4_country_prevalence.describe()
```

compute 95% uncertainty Interval

```python
# 2.5th and 97.5th percenttiles
uncertainty_interval = np.percentile(Overal_4_country_prevalence, [2.5, 97.5])
uncertainty_interval
```
